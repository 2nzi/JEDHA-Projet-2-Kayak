```python
import requests

def get_cities(city):
    url = "https://nominatim.openstreetmap.org/search"
    params = {
        "format": "json", #prendre geojson???
        "q": f"{city},France"
    }
    r = requests.get(url, params=params)
      
    return r.json()[0] 
     
get_cities('Nantes')
```
